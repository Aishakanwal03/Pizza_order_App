```python
available_size = ["Small", "Medium", "Large"]
available_flavour = ['Tikka' , 'Maxican', 'Fajita']
available_toppings = ['Olives' , 'Herbs', 'Extra Cheese']
print('welcome to pizza point')
your_order = []

while True:
    order = {}
    asked = input('Would you like to order pizza from pizza point? print "yes" for continue and print "no" for exit: ').title()
    if asked == 'Yes':
        print('Avaialable Menu')
        print('==Available size==')
        print(available_size)
        print()
        print('==Available flavour==')
        print(available_flavour)
        print()
        print('==Available toppings==')
        print(available_toppings)
        print()
        
        size = input('please enter the size of pizza you want to order: ').title()
        if size  in available_size:
            order['Size'] = size
            print('Thank you')
            print()
            
        elif size not in available_size:
            size = input('The size you enter is not available, please enter the size from the given size menu: ').title()
            if size in available_size:
                order['Size'] = size
                print('thank u')
                print()
            else:
                print('please login again and enter the correct spelling of size, as in given menu.')
                break
    else:
         break
            
    flavour = input('enter the flavour of pizza you want to order: ').title()
    if flavour in available_flavour:
        print('thank you')
        order['Flavour'] = flavour
        print()
        
    elif flavour not in available_flavour:
        flavour = input('The flavour you enter is not available, please enter the flavour from the given menu: ').title()
        if flavour in available_flavour:
            print('thank you')
            order['Flavour'] = flavour
            print()
        else:
            print('please login again and enter the correct spelling of flavour, as in given menu.')
            break
            
    
    topping = input('please enter the topping: ').title()
    if topping in available_toppings:
        print('thank you')
        order['Topping'] = topping
        print()
        
    elif topping not in available_toppings:
        topping = input('the topping you enter is not available, please enter the tooping from the given menu: ')
        if topping in available_toppings:
            print('thank you')
            order['Topping'] = topping
            print()
        else:
            print('please login again and enter the correct spelling of topping, as in given menu.')
            break
            
    your_order.append(order)
    
print()
print('your order is: ')    
print(your_order)
```
